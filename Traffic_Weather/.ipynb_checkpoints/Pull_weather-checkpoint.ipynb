# Now define functions to do this for us

```python
def pull_data(date):
    f = requests.get('http://api.wunderground.com/api/2bd9def127e30dea/history_'+date+'/q/TX/Austin.json')
    json_string = f.json()
    f.close()
    
    current_df = pd.DataFrame(json_string['history']['dailysummary'])
    a = current_df['date'].values.tolist()
    a = str(a)[1:len(str(a))-1]
    a = a.replace("'",'"')
    b = json.loads(a)
    c = pd.DataFrame.from_dict(b, orient='index')
    current_df = pd.concat([current_df, c.T], axis=1)
    
    return current_df
    
    
```

```python
def stack_df(old_df, new_df):
    stacked_df = pd.concat([old_df,new_df], axis=0)
    return stacked_df
```

```python
# Load dataframe with the dates 
dates_df = pd.read_csv(path+'Bluetooth_Travel_Sensors_-Traffic_Match_Summary_Records__TMSR_.csv', usecols=['timestamp'])
print(dates_df.shape[0])
dates_df.head()
```

```python
dates_df['timestamp'][0][:10]
```

```python
def get_day(x):
    return int(x[3:5])

def get_month(x):
    return int(x[0:2])

def get_year(x):
    return int(x[6:10])

def kill_time(x):
    return x[:10]
```

```python
dates_df['timestamp'] = dates_df['timestamp'].apply(lambda x: kill_time(x))
dates_df['Day'] = dates_df['timestamp'].apply(lambda x: get_day(x))
dates_df['Month'] = dates_df['timestamp'].apply(lambda x: get_month(x))
dates_df['Year'] = dates_df['timestamp'].apply(lambda x: get_year(x))
```

```python
dates_df = dates_df.loc[dates_df['Year']==2016]
dates_df = dates_df.drop_duplicates()
print(dates_df.shape[0])
dates_df.head()
```

```python
# Sort so dates are in order 
dates_df = dates_df.sort_values(by=['Month', 'Day'])
dates_df.head()
```

```python
def make_dates(x):
    date = x[6:10] + x[0:2] + x[3:5]
    return date
```

```python
dates_list = dates_df['timestamp'].apply(lambda x: make_dates(x))
len(dates_list)
```

```python
dates_list = dates_list.values.tolist()
```

```python
dlist = dates_list[1:] #already have 01/01/2016
```

# Run the loop

```python
df = pull_data('20160101')
df
```

```python
import time

i = 0
for d in dlist:
    temp_df = pull_data(d)
    df = stack_df(df, temp_df)
    i += 1
    if i == 10:
        i = 0
        print('Sleeping, last date pulled was', d)
        time.sleep(60) #API limit is 10 per min 

df
```
