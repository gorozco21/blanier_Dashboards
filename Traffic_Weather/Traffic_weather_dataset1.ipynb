# Some EDA on traffic

```python
# Change the time bin to be in line with the actual time 
def get_hour(t):
    hr = t[0:2]
    return int(hr)

def get_min(t):
    m = t[3:5]
    return int(m)
    
```

```python
traffic_df['TB_hour'] = traffic_df['Time Bin'].apply(lambda x: get_hour(x))
traffic_df['TB_minute'] = traffic_df['Time Bin'].apply(lambda x: get_min(x))

traffic_df.head()
```

```python
traffic_df['TB_hour'] = (traffic_df['TB_hour']+5)%24
# traffic_df.loc[traffic_df['Time Bin'] == '23:00']
```

```python
def convert_back(x):
    if x < 10:
        return '0'+str(x)
    return str(x)

def fix_tb(x):
    if len(x) < 5:
        return x+'0'
    return x
```

```python
traffic_df['TB_hour'] = traffic_df['TB_hour'].apply(lambda x: convert_back(x))
traffic_df['TB_minute'] = traffic_df['TB_minute'].apply(lambda x: str(x))

traffic_df['updated_Time_Bin'] = traffic_df['TB_hour'] + ':' + traffic_df['TB_minute']
traffic_df['updated_Time_Bin'] = traffic_df['updated_Time_Bin'].apply(lambda x: fix_tb(x))
traffic_df.head()
```

```python
def c_to_f(x):
    return int((x*1.8)+32)
```

```python
temp_df = weather_df[['DATE','TMAX', 'TMIN', 'PRCP']]
temp_df.head()
```

```python
temp_df['TMAX'] = temp_df['TMAX'].apply(c_to_f)
temp_df['TMIN'] = temp_df['TMIN'].apply(c_to_f)
temp_df.head()
```

```python
traf_df2 = traffic_df.copy()
```

```python
# Extract the date 
def parse_date_str(x):
    month = x[0:2]
    day = x[3:5]
    year = x[6:10]
    
    return day, month, year
```

```python
# Make a DATE col that matches the temp_df date col 

date_list = []

for i in range(traf_df2.shape[0]):
    day, month, year = parse_date_str(traffic_df['DATE'].iloc[i])
    date = year+'-'+month+'-'+day
    date_list.append(date)

# date_list
```

```python
traf_df2['DATE'] = date_list
traf_df2.head()
```

```python
traf_df2.shape[0], temp_df.shape[0]
```

```python
temp_df['DATE'][0]
```

```python
join_df1 = traf_df2.merge(temp_df, on='DATE')
join_df1['TAVG'] = (join_df1['TMAX'] + join_df1['TMIN']) / 2
join_df1.head()
```
